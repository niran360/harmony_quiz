```python
import numpy as np
import pandas as pd
import datasist as ds
```

```python
url = 'https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
df = pd.read_csv(url, index_col=0)
df.head()
```

```python
missing_value=df.isnull().sum()
missing_value
```

```python
lowest_avg_fuel_cost=df.groupby(["fuel_type_code_pudl"]).agg({"fuel_cost_per_unit_burned":"count"})
lowest_avg_fuel_cost
```

```python
percentile_75th=df.fuel_mmbtu_per_unit.quantile(0.75).round(2)
percentile_75th
```

```python
std_fuel_per_unit=df['fuel_mmbtu_per_unit'].round().std()
std_fuel_per_unit
```

```python
highest_fuel_cost_year=df.groupby(["report_year"]).agg({"fuel_cost_per_unit_delivered":"count"}).sort_values('fuel_cost_per_unit_delivered',ascending=False)
highest_fuel_cost_year
```
